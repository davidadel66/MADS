```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

# Load the mplo3td toolkit that is included with the Matplotlib.
from mpl_toolkits import mplot3d

#Three-dimensional plotting is one of the functionalities that benefits immensely from
# viewing figures interactively rather than statically in the notebook, so to use interactive
# figures, we can use%matplotlib notebook rather than %matplotlib inline

```

```python
# load a dataset
RentCali2 = pd.read_csv('../assets/RentCali2.csv')

RentCali2.head()
```

```python
# Now, we set up our figure and ax and pass it the keyword projection '3d' to create a 3-dimensional axes
fig =plt.figure()
ax = plt.axes(projection='3d')

```

```python
# Now let's create a 3d scatter plot of the national price, price and year. 
fig = plt.figure
ax = plt.axes(projection='3d')

# We can pass those rental cali variables x y and z
x = RentCali2['Year']
y = RentCali2['Price']
z = RentCali2['NationalPrice']
# In analogy with the more common two-dimensional plots, there can be created using the 
# ax.plot3D and ax.scatter3D functinos
ax.scatter3D(x,y,z)
# Add labels
ax.set_xlabel('Year')
ax.set_ylabel('OC Price')
ax.set_zlabel('National Price')

plt.show()
```

```python
# Notice that by default, the scatter poitns have their transperancy adjusted to give
# a sense of depth on the page.
# But overall this isn't really very informative data visualization.
# This is because two of our variable, year and average national price have limited variance,
# and as a result the 3rd dimension isn't really providing any new information.

# So this is a good example of a poor application of 3D plotting.
```

```python
# Another example
df = pd.DataFrame({'X': range(1,101),
                   'Y' : np.random.randn(100)*15+range(1,101),
                   'Z' : (np.random.randn(100)*15+range(1,101))})

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter3D(df['X'], df['Y'], df['Z'], c='skyblue', s=60)
ax.view_init(30,185) # This can be used to rotate the axes programmatically

ax.set_xlabel('kids')
ax.set_ylabel('Assistance')
ax.set_zlabel('Rent')

plt.show()
```
