```python
# Import the scripting layer, which is the interface through which users interact with matplotlib. It provides functinos and methods 
# to create and customize plots. 

import matplotlib.pyplot as plt
```

```python
# By default, Jupyter will change how matplotlib figures are rendered, and close figures at the end of cell execution. We will turn it off for now.
from IPython.display import set_matplotlib_close
set_matplotlib_close(False)
```

```python
# This initiated the scripting layer with a new figure where we can plot things. We can get a reference to this figure at any time
# by calling the get current figure, or gcf() function.
fig = plt.gcf()
fig
```

```python
# When a figure is created in matplotlib it is also given an identifier. We can ask the pyplot scripting layer to tell us all of the figures which currently exist by looking at the figure numbers. 
plt.get_fignums()
```

```python
# Get the figure axes object
fig.get_axes()
```

```python
# Or you can use gca function to get the current axes
plt.gca()
```

```python
# Example
plt.plot([1,2,3], [4,6,10])
```

```python
# Show the plot
plt.show()
```

```python
# Matplotlib did some work for us by connecting the points by default, and drawing axis objects and setting the scale for our plot.

# Remember there are two axis object per axes, the x axis and the y axis. We can get these objects directly if we want to

x=plt.gca().get_xaxis()
y=plt.gca().get_yaxis()

# Once we have the Axis object we an inspect them and change them. 
x.set_label_text("Numbers one through three")
y.set_label_text("Up to ten")

plt.show()
```

```python
 # Add in more data
plt.plot([1,22,3], [4,6,10])
plt.show()
```
