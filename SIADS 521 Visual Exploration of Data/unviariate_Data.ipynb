```python
# Import the necessary libraries
import pandas as pd
import numpy as np

# Load the Iris dataset
iris = pd.read_csv('./assets/iris.csv')

print(iris.info())

iris.head()
```

```python
# The .mean function will give us the mean of a single variable 
iris['sepal_length'].mean()
```

```python
# Calculate the standard deviation
iris['sepal_length'].std()
```

```python
# However, when exploring our data we usally want to know a bit more about it.
# So, lets use pandas describe function to calculate the mean, std, interquartile range IQR values for sepal_length

iris['sepal_length'].describe()
```

```python
# Do it over all columns
iris.describe()
```

```python
# To get the summary statistics or descriptive statistics of both numeric and character column
iris.describe(include='all')
```

```python
# Remember that there are 3 different species in the dataset. So we would want to explore summary statistics by each group.
# To do this, we can groupby

iris.groupby('species')['petal_length'].mean()
```

```python
iris.groupby('species').describe()
```

```python
# Now lets explore the data visualization

import matplotlib as mpl
import matplotlib.pyplot as plt

```

```python
# Calculate mean and Std of each variable

# Mean
sepal_length_mean = iris['sepal_length'].mean()
petal_length_mean = iris['petal_length'].mean()
sepal_width_mean = iris['sepal_width'].mean()
petal_width_mean = iris['petal_width'].mean()

# Std Deviation
sepal_lengthstd = iris['sepal_length'].std()
petal_lengthstd = iris['petal_length'].std()
sepal_widthstd = iris['sepal_width'].std()
petal_widthstd = iris['petal_width'].std()


```

```python
# Lets Create a bar graph of the values
plt.bar(['sepal_length', 'petal_length', 'pepal_width', 'setal_width'], [sepal_length_mean, petal_length_mean, petal_width_mean, sepal_length_mean])
plt.show()
```

```python
iris_mean = iris.groupby('species').mean()
iris_std = iris.groupby('species').std()

print(iris_std)
print(iris_mean)
```

```python
h_groups = 4 
means_setosa = iris_mean.loc['setosa'].values
std_setosa = iris_std.loc['setosa'].values

means_versicolor = iris_mean.loc['versicolor'].values
std_versicolor = iris_std.loc['versicolor'].values

means_virginica = iris_mean.loc['virginica'].values
std_virginica = iris_std.loc['virginica'].values


# Create the plot
# We will create a figure object called fig and an axis object in that figure called ax using plt.subplots function
fig, ax = plt.subplots()

index = np.arange(h_groups)
bar_width = 0.25

# Adjust the transperancy of the bars
opacity = 0.4

# error bar
error_config = {'ecolor': '0.3'}

# Put a bar chart on our ax with the ax.bar() method
rectsl = ax.bar(
    index,
    means_setosa,
    bar_width,
    alpha=opacity, color ='b',
    yerr=std_setosa, error_kw=error_config,
    label='Setosa'
)

rects2 = ax.bar(
    index + bar_width,
    means_versicolor,
    bar_width,
    alpha=opacity, color ='r',
    yerr=std_versicolor, error_kw=error_config,
    label='Versicolor')

rects3 = ax.bar(
    index + bar_width + bar_width,
    means_virginica,
    bar_width,
    alpha=opacity, color ='c',
    yerr=std_virginica, error_kw=error_config,
    label='Virginica'
)


ax.set_xlabel('Variable')
ax.set_ylabel('Means')
ax.set_title('Means by variable and Species')
ax.set_xticks(index + bar_width/3)
ax.set_yticklabels(('sepal length', 'sepal width', 'petal length', 'petal width'))
ax.legend()

fig.tight_layout()
plt.show()

```
