```python
# import the necessary libraries

import pandas as pd
import matplotlib as mpl
mpl.get_backend()
import matplotlib.pyplot as plt

# Load the iris dataset
iris = pd.read_csv('./assets/iris.csv')


# Print the first 5 rows of iris
print(iris.head())
```

```python
# We can use pandas .quantile method to calculate 3 quartiles, the lower 25%, 50% and the upper 25% for the sepal_length variable

iris['sepal_length'].quantile([.25, .5, .75])
```

```python
# The best way to chart the data is to use a boxplot. 
# The five cariables of the boxplot are, the minimum values, 25th percentile, median, 75th percentile and the maximum value. The mean isnt included

# Plot using the boxplot function in the plt library, and we will need to pass in an array or a list of vectors

plt.boxplot(iris['sepal_length'])
plt.show()
```

```python
# Confirm the numbers using th describe function

iris['sepal_length'].describe()
```

```python
# We can also have the boxplot show horizontally

plt.boxplot(iris['sepal_length'], showmeans=True, vert=False)

plt.show()
```

```python
# Lets do more than one boxplot

irisvars=[iris['sepal_length'], iris['sepal_width'], iris['petal_length'], iris['petal_width']]
plt.boxplot(irisvars, showmeans=True)
plt.show
```

```python
# Each boxplot can also have its own color. we can dop so by modifying the color in the setp() function in pyplot.

box = plt.boxplot(irisvars, showmeans=True)

plt.setp(box['boxes'][0], color='red')
plt.setp(box['caps'][0], color='red')
plt.setp(box['whiskers'][0], color='red')
```
