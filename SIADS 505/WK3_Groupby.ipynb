In this file, we will explore all the different options to groupby in rows and columns, etc of a DataFrame. 

## Splitting

```python
# Import the important packages

import numpy as np
import pandas as pd
```

```python
# Read the census data
df = pd.read_csv('./resources/census.csv')

# Exclude state level summarizations, which have sum  level of 40
df =df[df['SUMLEV'] == 50 ]
df.head()
```

```python
# In the first example for groupby() I want to use the census date. Let's get a list of the unique states,
# then we can iterate over all the states and for each state we reduce the data frame and calculate the average. 

for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('Counties in state ' + state + ' have an average population of ' + str(avg))
```

```python
# Let's do the same result by using the groupb() approach

for group, frame in df.groupby('STNAME'):
    # We set two values here. Groupby() returns a tuple, where the first values is the
    # value of the key we were trying to group by, in this case a specific state name, and the second one is the 
    # projected dataframe that was found for that group

    # Now we include our logic in the "apply" step, which is to calculate an average of the census2010pop
    avg = np.average(frame['CENSUS2010POP'])
    # Print the results
    print('Counties in state ' + group + 'have an average population of ' + str(avg))

    # This method is much quicker than the first one
```

```python
# You can also provide a function to group by and use that to segment your data.

# Say you want to work on third or so of the states at  a given time. We could create some function which returns a number 
# between zero and two based on the first character of the state name. Then we can tell group by to use this function to split up our data frame.
# Note that in order to do this, you need to set the index of the df to be the column of what you want to group by

df = df.set_index('STNAME')

def set_batch_number(item):
    if item[0]<'M' :
        return 0
    elif item[0]<'Q':
        return 1
    else:
        return 2
    
for group, frame in df.groupby(set_batch_number):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing')
```

```python
# Another example
df = pd.read_csv('./resources/listings.csv')
df.head()
```

```python
# Say there are two columns of interest, one is cancellation_policy and the other is review_scores_value

df = df.set_index(['cancellation_policy', 'review_scores_value'])

# When we have a multiindex we need to pass in the levels we are interested in grouping by
for group, frame in df.groupby(level=(0,1)):
    print(group)

```

```python
# This seems to work ok, but what if wanted to group by the cancelation policy and the review scores, but 
# separate out all the 10's from those under ten? In this case, we could use a function to manage the groupings

def grouping_fun(item):
    # Check the review_scores_value portion of the idx
    if item[1] == 10.0:
        return (item[0], "10.0")
    else:
        return (item[0], "not 10.0")

for group, frame in df.groupby(by=grouping_fun):
    print(group)
    
```

```python
 # To this point we have applied very simple processing to our data after splitting, really just outputting some print statements
 # to demonstrate how the splitting works. Pandas developers have three broad categories of data processing to happen during the apply step
 # Aggregation of group data, Transformation of group data,  and Filteration of group data
 
```

## Aggregation

```python
# The most straight forward apply step is the aggregation of data. it uses the agg() method on the groupby object
# With agg we can pass in a dictionary of the columns we are interested in aggregating along with the function we are looking
# to apply to aggregate

# Reset the index of the airbnb data
df = df.reset_index()

# Now let's group by the cancellation policy and find the average review_scores_value by group
df.groupby('cancellation_policy').agg({"review_scores_value":np.average})
```

```python
# np.avrage does not ignore nans! However, there is a function we can use for this
df.groupby("cancellation_policy").agg({"review_scores_value":np.nanmean})
```

```python
# We can just extend this dictionary to aggregate by multiple functions or multiple columns
df.groupby("cancellation_policy").agg({"review_scores_value":(np.nanmean, np.nanstd),"reviews_per_month": np.nanmean})
```

```python
# In the last cell, first we're doing a group by on the dataframe object by the column "cancellation_policy". This creates a new GroupBy object.
# Then we are invoking the agg() function on that object. The agg function is going to apply one or more functions we specify to the group dataframes
# and return a single orw per dataframe/group. When we called this function we sent it two dictionary entries, each with the key indicating which column
# we wanted functions applied to. For the first column we supplied tuple of two functions.
# These are not function invocations like np.nanmean(), by they are references to functions which will return single values.
```

## Transformation

```python
# Agg returns a single value per column, so one row per group. transform() returns an object that is the same size
# as the group. Essentially it broadcasts thefunction you supply over the grouped dataframe, returning a new dataframe.

# For instance ,suppose we want to include the average rating values in a given group by cancellation policy,
# but preserve the dataframe shape so that we could generate a differnece between an individual observation and the sum
```

```python
# First, lets define just some subset of columns we are interested in 
cols = ['cancellation_policy' , 'review_scores_value']
# Now lets transform it, I'll store this in its own dataframe
transform_df = df[cols].groupby('cancellation_policy').transform(np.nanmean)
transform_df.head()
```

```python
# The index is the same as the original df. SO lets just join this in.
transform_df.rename({'review_scores_value': 'mean_review_scores'}, axis ='columns', inplace=True)
df = df.merge(transform_df,left_index=True, right_index=True)
df.head()
```

```python
# We can see our new column is in place. the mean_review_scores. So now we could create, for instance, the difference between a given row and its group (cancellation policy) means
df['mean_diff'] = np.absolute(df['review_scores_value'] - df['mean_review_scores'])
df['mean_diff'].head()
```

## Filtering

```python
# The groupby object has build in support for filtering groups as well. It's often that you'll want to group by some feature
# then make some transformatino to the groups, then drop certain groups as part of your cleaning routines.
# filter() function takes in a function which it applies to each group dataframe and returns either a True or False
# depending on whether that group should be included in the results

# For instance, if we only want those groups which have a mean rating above 9 included in our results
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value']) > 9.2)
```

```python
# The results are still indexedm by that any of the results which were in a gorup with a mean review score of less than or equal to 9.2 were not copied over
```

## Applying

```python
# By far the most common operation I invoke on groupby objects is the appl() function. This allows you to apply an arbitrary function to each group
# and stitch the results back for each apply() into a single dataframe where the index is preserved

# Lets look at an example using our airbnb data. 
df = pd.read_csv('./resources/listings.csv')
# Include some of the columns of interest
df = df[['cancellation_policy','review_scores_value']]
df.head()
```

```python
# In preview work we wanted to find the average review score of a lisuting and tis deviation from the group mean.
# This was a two step process, first, we used transform() on the groupby object and then we had to broadcast to create a new column.
# With apply() we could wrap this logic in one place
def calc_mean_review_scores(group):
    avg = np.nanmean(group['review_scores_value'])
    # broadcast our formula and create a new column
    group['review_scores_mean']=np.abs(avg-group['review_scores_value'])
    return group

# Now we apply this to the groups
df.groupby('cancellation_policy').apply(calc_mean_review_scores).head()
```
